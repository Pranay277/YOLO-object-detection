```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os

input_folder = r"D:\pranay cse\IITh\img"
output_folder = r"D:\pranay cse\IITh\Outputs"


os.makedirs(output_folder, exist_ok=True)
model = YOLO("yolo11m.pt")
results = model.predict(
    source=input_folder,
    imgsz=640,
    conf=0.25,
    save=True,               
    save_dir=output_folder   
)
plt.figure(figsize=(15, 8))
for i, r in enumerate(results):    
    im_array = r.plot()
    im_rgb = cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)
    plt.subplot(1, len(results), i + 1)
    plt.imshow(im_rgb)
    plt.axis("off")
    plt.title(f"Image {i+1}")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

base_path = r"D:\pranay cse\IITh\Task_2\runs\detect\val4"

# Define files that actually exist in your folder
conf_matrix_path = f"{base_path}\\confusion_matrix.png"
conf_matrix_norm_path = f"{base_path}\\confusion_matrix_normalized.png"
box_f1_curve = f"{base_path}\\BoxF1_curve.png"
box_p_curve = f"{base_path}\\BoxP_curve.png"
box_pr_curve = f"{base_path}\\BoxPR_curve.png"
box_r_curve = f"{base_path}\\BoxR_curve.png"

def show_image(path, title):
    """Read and display an image using OpenCV + Matplotlib"""
    img = cv2.imread(path)
    if img is None:
        print(f" Could not read {path}")
        return
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8,6))
    plt.imshow(img)
    plt.axis("off")
    plt.title(title)
    plt.show()

# Display all available evaluation plots
show_image(conf_matrix_path, "Confusion Matrix")
show_image(conf_matrix_norm_path, "Normalized Confusion Matrix")
show_image(box_f1_curve, "F1 Curve")
show_image(box_p_curve, "Precision Curve")
show_image(box_pr_curve, "Precision-Recall Curve")
show_image(box_r_curve, "Recall Curve")
```
